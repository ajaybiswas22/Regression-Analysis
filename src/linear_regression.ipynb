In this assignment, we have to predict house prices and measure prediction error using linear regression. Linear regression is a linear model that assumes a linear relationship between the input variables (x) and the single output variable (y). We will be drawing a straight line which fits all the points and minimizes the error.

```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
```

### Load Dataset

```python
# load csv
csvfile = pd.read_csv("Real estate valuation data set.csv") 
print(csvfile)

# loading dataset into an array
dataset = csvfile.values

# Loading real House price in Y
Y = dataset[:,6]
```

### Regression Constant and Regression Coefficient

> Regression Constant  $w1 = \frac{\Sigma y (\Sigma x^2) - (\Sigma x)(\Sigma xy)}{n\Sigma x^2 - (\Sigma x)^2} $

> Regression Coefficient $w2 = \frac{\Sigma x \Sigma y - n \Sigma xy}{(\Sigma x)^2 - n\Sigma x^2} $


### Regression Line

> $\hat y = w_1 + w_2x$

where $\hat y $ is the dependable variable, and $x$ is the independant variable.


```python
def corr_const_and_coeff(x,y):
    
    sum_y = y.sum()
    sum_x = x.sum()
    sum_x_sq = (x**2).sum()
    sum_xy = (x*y).sum()
    sum_y_x_wq = (y*(x**2).sum()).sum()
    n = len(x)
    
    w1 = ( sum_y_x_wq - (sum_x * sum_xy) ) / ( (n * sum_x_sq) - sum_x**2 )
    w2 = ( ( sum_x * sum_y ) - (n * sum_xy) ) / ( sum_x**2 - (n * sum_x_sq) )
    
    return w1,w2
```

### Predicting house price based on house age

```python
X1 = dataset[:,1]

plt.plot(X1,Y,'*')
plt.xlabel('House Age (X1)')
plt.ylabel('House Price (Y)')
plt.title('Predicting House Price Based on House Age')

w1,w2 = corr_const_and_coeff(X1,Y)

y_hat1 = np.array([w1 + w2 * value for value in X1 ])

plt.plot(X1, y_hat1, linestyle='-')
plt.legend(["Data Point","Fitted Regression line"])
```

### Predicting house price based on the distance to the nearest MRT station

```python
X2 = dataset[:,2]

plt.plot(X2,Y,'*')
plt.xlabel('Distance to the nearest MRT station (X2)')
plt.ylabel('House Price (Y)')
plt.title('Predicting House Price Based on the distance to the nearest MRT station')

w1,w2 = corr_const_and_coeff(X2,Y)

y_hat2 = np.array([w1 + w2 * value for value in X2 ])

plt.plot(X2, y_hat2, linestyle='-')
plt.legend(["Data Point","Fitted Regression line"])
```

### Calculating Error

```python
e1 = Y - y_hat1
e2 = Y - y_hat2
e3 = Y - y_hat3
e4 = Y - y_hat4
e5 = Y - y_hat5
```

### Error Tables

```python
# error table

# Table 1
e_tab1 = np.zeros((len(Y),4))
e_tab1[:,0] = np.arange(1,len(Y)+1,dtype=int)
e_tab1[:,1] = Y
e_tab1[:,2] = y_hat1
e_tab1[:,3] = e1

# Table 2
e_tab2 = np.zeros((len(Y),4))
e_tab2[:,0] = np.arange(1,len(Y)+1)
e_tab2[:,1] = Y
e_tab2[:,2] = y_hat2
e_tab2[:,3] = e2

# Table 3
e_tab3 = np.zeros((len(Y),4))
e_tab3[:,0] = np.arange(1,len(Y)+1)
e_tab3[:,1] = Y
e_tab3[:,2] = y_hat3
e_tab3[:,3] = e3

# Table 4
e_tab4 = np.zeros((len(Y),4))
e_tab4[:,0] = np.arange(1,len(Y)+1)
e_tab4[:,1] = Y
e_tab4[:,2] = y_hat4
e_tab4[:,3] = e4

# Table 5
e_tab5 = np.zeros((len(Y),4))
e_tab5[:,0] = np.arange(1,len(Y)+1)
e_tab5[:,1] = Y
e_tab5[:,2] = y_hat5
e_tab5[:,3] = e5

# Dataframes
et1 = pd.DataFrame(e_tab1, columns = ['Standard Order', 'Actual Value(Y)', 'Fitted Value(y_hat1)','Residual'])
et2 = pd.DataFrame(e_tab2, columns = ['Standard Order', 'Actual Value(Y)', 'Fitted Value(y_hat2)','Residual'])
et3 = pd.DataFrame(e_tab3, columns = ['Standard Order', 'Actual Value(Y)', 'Fitted Value(y_hat3)','Residual'])
et4 = pd.DataFrame(e_tab4, columns = ['Standard Order', 'Actual Value(Y)', 'Fitted Value(y_hat4)','Residual'])
et5 = pd.DataFrame(e_tab5, columns = ['Standard Order', 'Actual Value(Y)', 'Fitted Value(y_hat5)','Residual'])

print('\nError Table for Linear Regression using X1\n')
print(et1)

print('\nError Table for Linear Regression using X2\n')
print(et2)

print('\nError Table for Linear Regression using X3\n')
print(et3)

print('\nError Table for Linear Regression using X4\n')
print(et4)

print('\nError Table for Linear Regression using X5\n')
print(et5)

# Saving tables to output folder
et1.to_csv (r'..\output\table1.csv', index = False, header=True)
et2.to_csv (r'..\output\table2.csv', index = False, header=True)
et3.to_csv (r'..\output\table3.csv', index = False, header=True)
et4.to_csv (r'..\output\table4.csv', index = False, header=True)
et5.to_csv (r'..\output\table5.csv', index = False, header=True)
```
